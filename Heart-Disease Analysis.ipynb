```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("../Pandas/heart-disease.csv")
heart_disease
```

```python
#Sample Histogram of the 'age' in following dataset 

heart_disease["age"].plot.hist(bins=45);
```

```python
#This would be better viewed if we created subplots of histograms of all columns given in daya

heart_disease.plot.hist(figsize=(15,40), subplots=True, bins=30);
```

```python
#There is a lot of information given. We can target patients who has age of more than 50. 
#Let's create file of those who has age >50 from heart_dieses file

over_50 = heart_disease[heart_disease["age"]>50]
over_50
```

```python
# Let's Create subplot of 3 entities chol,age & thalach. This visualization is from the new created dataset "over_50" 
#Here thalach means: person's maximum heart rate
# Chol means: Cholestrol levels of patient's body
# & age: Patient's age at time

#For Customization I have used 'seaborn-whitegrid' plot style of matplotlib.
plt.style.use('seaborn-whitegrid')

fig, ((ax0,ax1)) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(17,15));

#Let's set data for ax0 and ax1 now:

#For ax0:
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter") #For the color theme
#Let's Custom ax0:
ax0.set(title="Heart Disease & Cholestrol levels",
        xlabel="Age",
        ylabel="Cholestrol");
#Changes ax0 axes limit:
ax0.set_xlim([50,80])
#Adding legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");
#Adding a meanline to compare patient data in plot:
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");


#Let's set data to ax1:
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter") #For the color theme
#Let's Custom ax1:
ax1.set(title="Heart Disease & Cholestrol levels",
        xlabel="Age",
        ylabel="Thalach");
#Changes ax1 axes limit:
ax1.set_xlim([50,80])
#Adding legend to ax1:
ax1.legend(*scatter.legend_elements(), title="Target");
#Adding a meanline to compare patient data in plot:
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");


#Adding title for figures 
fig.suptitle("Heart Disease Analyses", fontsize=20, fontweight="bold");
```

```python
#Saving the plot 
fig.savefig("Heart-Disease Analysis.png");
```
